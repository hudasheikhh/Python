## Drop Rows From DataFrame

By using pandas.DataFrame.drop() method you can drop/remove/delete rows from DataFrame. axis param is used to specify what axis you would like to remove. By default axis = 0 meaning to remove rows. Use axis=1 or columns param to remove columns. By default, Pandas return a copy DataFrame after deleting rows, used inpalce=True to remove from existing referring DataFrame.
DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')

```python
df_copy.drop(index=[1,2]) #inplace is False so this command won't afffect the actual DataFrame instead it returns the filtered output
```

```python
df_copy.drop(index=[0,1],axis=0)   #Delete row by index
```

```python
df_copy.drop(df_copy.index[0:3])   ## Delete Rows by Index Range
df_copy.drop(range(0,3))            # Delete row using range function
```

```python

# Iterate all rows using DataFrame.iterrows()
print("After iterating all rows:\n")

for index, row in df_copy.iterrows():
    print (f"{index} {(row["first_name"]).ljust(15," ")} paid {row["amount"]} on {row["payment_date"]}") #The ljust method means "left-justify"; this makes sense because your string will be to the left after adjustment up to the specified length.
```
