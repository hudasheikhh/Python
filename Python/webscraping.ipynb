```python
from bs4 import BeautifulSoup

import requests
web_url=requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=Python+&txtLocation=mumbai").text
soup = BeautifulSoup(web_url,'lxml') 

# h2_tag = soup.find_all('h2')

li_tag = soup.find_all('li', class_='clearfix job-bx wht-shd-bx')
print(li_tag)
```
